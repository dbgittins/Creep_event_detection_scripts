```python
import pandas as pd
import datetime as dt
```

```python
Creepmeter_dataframe = pd.read_csv('../../Data/creepmeter_lookup.csv',index_col=0)

creepmeter = 'C461'

creepmeter_info = Creepmeter_dataframe.loc[Creepmeter_dataframe.index[Creepmeter_dataframe.Creepmeter_abbrv==creepmeter]]

creep_phases = pd.read_csv('../../Gittins_and_Hawthorne_event_catalogue/Creep_phases_{k}.csv'.format(k=creepmeter[0:3]),index_col=0)

Creep_catalogue = pd.read_csv('../../Gittins_and_Hawthorne_event_catalogue/Creep_event_catalog_at_{k}_Parkfield_and_Rain.csv'.format(k=creepmeter[0:3]),index_col=0)

Start_times = []

for i in range(len(Creep_catalogue)):
    row = creep_phases.iloc[i]
    row.dropna(inplace=True)
    for j in range(int((len(row)/2))-1):
        shift = row['T0{k}'.format(k=j+1)]-2
        Start_times.append(pd.to_datetime(Creep_catalogue.Start_Time.iloc[i])+dt.timedelta(hours=shift))
Start_times = pd.to_datetime(Start_times)
df_picks=pd.DataFrame(Start_times,columns=['ST'])
df_picks.to_csv("../../../Data/{p}/Picks/{q}_picks.csv".format(p=creepmeter_info.Network.iloc[0],q =creepmeter_info.File_code.iloc[0]))

```
