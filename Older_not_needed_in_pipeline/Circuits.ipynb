```python
import schemdraw
import schemdraw.elements as elm

# Initialize the drawing
d = schemdraw.Drawing()

# Power source and regulator section
d += elm.SourceV().up().label('12V').at((0,0))
d += elm.Line().down(d.unit/2)
d += elm.Resistor().down().label('Voltage Regulator')
d += elm.Line().right(d.unit/2).label('5V').dot()

# Hall Sensor 1
d += elm.Line().down(d.unit*1.5).label('Hall Sensor 1')
d += elm.Line().right(d.unit/2)
d += elm.Opamp().label('Buffer')
d += elm.Line().right(d.unit/2)
d += elm.Dot()
d += elm.Line().right(d.unit/2).label('Output to Logger Ch1', 'center')

# Hall Sensor 2
d += elm.Line().down(d.unit*1.5).label('Hall Sensor 2')
d += elm.Line().right(d.unit/2)
d += elm.Opamp().label('Buffer')
d += elm.Line().right(d.unit/2)
d += elm.Dot()
d += elm.Line().right(d.unit/2).label('Output to Logger Ch2', 'center')

# Pulse detection circuit
d += elm.Line().down(d.unit*2.5).label('Pulse from Logger (1 ms)')
d += elm.Line().right(d.unit/2)
d += elm.MosfetN().label('Pulse Detection', 'top')
d += elm.Line().right(d.unit/2)
d += elm.Dot().label('Control Signal')

# Show drawing
d.draw()
```
